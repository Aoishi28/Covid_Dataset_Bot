```python
from selenium import webdriver
import time
import pandas as pd
from datetime import date
```

```python
chrome_browser = webdriver.Chrome()
chrome_browser.get("https://www.mohfw.gov.in/")
time.sleep(15)
```

```python
# To click on the expand button to load the table
exp=chrome_browser.find_element_by_xpath('//*[@id="state-data"]/div/div/div/h2/a')
exp.click()
```

```python
today=date.today()
today_date = str(today) # Since it was a datetime object we convert in into string
```

```python
head_list=[]
h= chrome_browser.find_element_by_xpath('//*[@id="state-data"]/div/div/div/div/table') #To extract the table only
th_list=h.find_elements_by_tag_name('th') #Extracting the headings


for i in th_list:
    head_list.append(i.text)
    
print(head_list)
print(len(head_list))
```

```python
column_names=['S. No.', 'Name of State / UT', 'Active_Cases_Total', 'Active_Cases_Change since yesterday', 'Cured/Discharged/Migrated_Cumulative', 'Cured/Discharged/Migrated_Change since yesterday', 'Deaths_Cumulative', 'Deaths_Change since yesterday']
df=pd.DataFrame(columns= column_names) #Creating a dataframe with the proper headings
print(df)
```

```python
tr_list=h.find_elements_by_tag_name('tr') # Each tr corresponds to a particular state

for i in tr_list:
    td_list=i.find_elements_by_tag_name('td') # td retrieve each piece of info for a state
    row=[]
    for td in td_list:
        row.append(td.text) # creating row i.e. each state data
    print(row)
    
    try:
        data = {}
        for j in range(len(df.columns)):
            data[df.columns[j]] = row[j] #Creating a dictionary to be appended to the dataframe
        df = df.append(data, ignore_index=True) #Appending the data to the dataframe
    except:
        continue #Some extra lists are returned which would have resulted in error thus putting it in except block
    
    
```

```python
df=df.drop(columns=['S. No.']) #Removing the S. No. column since it doesn't convey any information
```
